```python
def simulate_improvement(
    match,
    feature,
    increments
):
    
    results = []

    for inc in increments:

        modified = match.copy()

        modified[feature] += inc

        prob = predict_win_probability(
            modified[features]
        )

        results.append({
            "improvement": inc,
            "probability": prob
        })

    return pd.DataFrame(results)
```

```python
def feature_impact_ranking(
    match
):
    
    baseline = predict_win_probability(
        match[features]
    )

    impacts = []

    for feature in [
        "first_serve_win_pct",
        "second_serve_win_pct",
        "bp_save_pct"
    ]:

        modified = match.copy()

        modified[feature] += 0.05

        new_prob = predict_win_probability(
            modified[features]
        )

        impacts.append({
                "feature": feature,
                "gain":
                new_prob - baseline
        })

    return (
        pd.DataFrame(impacts)
        .sort_values(
            "gain",
            ascending=False
        )
    )
```

```python
loss_matches = model_df[
    model_df["result"] == 0
]
```

```python
def average_feature_gain(
    feature,
    improvement=0.05
):
    
    gains = []

    for idx in loss_matches.index:

        match = loss_matches.loc[[idx]]

        baseline = predict_win_probability(
            match[features]
        )

        modified = match.copy()

        modified[feature] += improvement

        new_prob = predict_win_probability(
            modified[features]
        )

        gains.append(
            new_prob - baseline
        )

    return np.mean(gains)
```

```python
season_results = []

for feature in [
    "first_serve_win_pct",
    "second_serve_win_pct",
    "bp_save_pct"
]:

    gain = average_feature_gain(feature)

    season_results.append({
        "feature": feature,
        "avg_gain": gain
    })

season_results = pd.DataFrame(
    season_results
)

season_results.sort_values(
    by="avg_gain",
    ascending=False
)
```
